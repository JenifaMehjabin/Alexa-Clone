```python
import pyttsx3
import speech_recognition as sr

# Initialize the pyttsx3 engine for TTS
engine = pyttsx3.init()

# Function for TTS
def speak(text):
    engine.say(text)
    engine.runAndWait()

# Initialize the recognizer for Speech-to-Text
recognizer = sr.Recognizer()

# Function to listen and recognize speech
def listen():
    with sr.Microphone() as source:
        print("Listening...")
        audio = recognizer.listen(source)
    try:
        text = recognizer.recognize_google(audio)
        print(f"Recognized: {text}")
        return text.lower()
    except sr.UnknownValueError:
        speak("Sorry, I couldn't understand that.")
        return ""
    except sr.RequestError:
        speak("Sorry, there was an error with the speech recognition service.")
        return ""
```

```python
import datetime

def get_time():
    return datetime.datetime.now().strftime("%H:%M:%S")

def get_date():
    return datetime.datetime.now().strftime("%A, %d %B %Y")

# Example command: "What time is it?"
def respond_to_time_command(command):
    if "time" in command:
        speak(f"The time is {get_time()}")
    elif "date" in command:
        speak(f"Today's date is {get_date()}")
```

```python
import webbrowser

def search_web(query):
    webbrowser.open(f"https://www.google.com/search?q={query}")
    speak(f"Searching for {query}")
```

```python
import os

def open_application(app_name):
    if "chrome" in app_name:
        os.system("start chrome")
    elif "notepad" in app_name:
        os.system("start notepad")
    else:
        speak("Sorry, I can't open that application.")

# Example command: "Open Chrome"
def handle_app_command(command):
    if "open" in command:
        app_name = command.split("open", 1)[1].strip()
        open_application(app_name)
```

```python
import requests

def get_weather(city):
    api_key = "your_openweathermap_api_key"  # Replace with your API key
    base_url = "http://api.openweathermap.org/data/2.5/weather?"
    complete_url = f"{base_url}q={city}&appid={api_key}"
    
    response = requests.get(complete_url)
    data = response.json()

    if data["cod"] != "404":
        main = data["main"]
        weather_desc = data["weather"][0]["description"]
        temperature = main["temp"] - 273.15  # Convert from Kelvin to Celsius
        speak(f"The weather in {city} is {weather_desc} with a temperature of {temperature:.2f}°C.")
    else:
        speak("City not found.")

# Example command: "What's the weather in London?"
def handle_weather_command(command):
    if "weather" in command:
        city = command.split("in", 1)[1].strip()
        get_weather(city)
```

```python
import smtplib

def send_email(subject, body, to_email):
    # Set up the email
    from_email = "your_email@gmail.com"  # Replace with your email
    password = "your_email_password"  # Use an app password for Gmail
    
    server = smtplib.SMTP("smtp.gmail.com", 587)
    server.starttls()
    server.login(from_email, password)
    
    message = f"Subject: {subject}\n\n{body}"
    server.sendmail(from_email, to_email, message)
    server.quit()
    speak("Your email has been sent.")
    
# Example command: "Send an email to John saying 'Hello!'"
def handle_email_command(command):
    if "send email" in command:
        subject = "Hello!"
        body = "Hello, this is your voice assistant."
        to_email = "recipient_email@example.com"  # Replace with recipient's email
        send_email(subject, body, to_email)
```

```python
import pyjokes

def tell_joke():
    joke = pyjokes.get_joke()
    speak(joke)

# Example command: "Tell me a joke"
def handle_joke_command(command):
    if "joke" in command:
        tell_joke()
```

```python
def handle_command(command):
    if "open powerpoint" in command or "launch powerpoint" in command:
        open_application("PowerPoint")
    elif "how are you" in command:
        speak("I'm doing great, thank you!")
    elif "exit" in command:
        speak("Goodbye!")
        return False
    else:
        speak("Sorry, I couldn't understand that.")
    return True
```

```python
def listen():
    with sr.Microphone() as source:
        print("Listening...")
        audio = recognizer.listen(source)
    try:
        # Try Google's speech recognition API
        text = recognizer.recognize_google(audio)
        print(f"Recognized: {text}")
        return text.lower()
    except sr.UnknownValueError:
        # Fallback to offline recognizer if Google doesn't recognize
        print("Google Speech Recognition could not understand audio, trying Sphinx...")
        text = recognizer.recognize_sphinx(audio)
        print(f"Recognized: {text}")
        return text.lower()
    except sr.RequestError:
        speak("Sorry, I couldn't process your request.")
        return ""
```

```python
def handle_weather_command(command):
    if "weather" in command:
        # Check if "in" is part of the command to get the city
        if "in" in command:
            city = command.split("in", 1)[1].strip()  # Get the city name after "in"
            get_weather(city)
        else:
            speak("Please provide a city for the weather.")
```

```python
def process_command(command):
    keywords = {
        "powerpoint": open_application,
        "chrome": open_application,
        "email": send_email,
        "weather": get_weather,
        "time": get_time
    }

    for key in keywords:
        if key in command:
            keywords[key](command)
            break
    else:
        speak("Sorry, I couldn't understand that.")
```

```python
import speech_recognition as sr

# Initialize recognizer
recognizer = sr.Recognizer()

# Use the microphone
with sr.Microphone() as source:
    print("Adjusting for ambient noise...")
    recognizer.adjust_for_ambient_noise(source)
    print("Say something...")
    audio = recognizer.listen(source)

# Recognize speech using PocketSphinx
try:
    print("Recognizing...")
    text = recognizer.recognize_sphinx(audio)
    print("You said: " + text)
except sr.UnknownValueError:
    print("PocketSphinx could not understand the audio")
except sr.RequestError as e:
    print(f"Error: {e}")
```

```python
while True:
    speak("How can Alexa assist you?")
    command = listen()
    
    if "exit" in command or "quit" in command:
        speak("Goodbye!")
        break
    elif "time" in command or "date" in command:
        respond_to_time_command(command)
    elif "search" in command:
        query = command.split("search", 1)[1].strip()
        search_web(query)
    elif "open" in command:
        handle_app_command(command)
    elif "weather" in command:
        handle_weather_command(command)
    elif "email" in command:
        handle_email_command(command)
    elif "joke" in command:
        handle_joke_command(command)
    else:
        speak("Sorry, I didn't understand that.")
```
